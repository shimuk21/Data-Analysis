## Exploratory Analysis and Visualization

**Here we will use the above data to analyse using Pandas and visualise the data using matplotlib and seaborn**



Let's begin by importing`matplotlib.pyplot` and `seaborn`.

```python
!pip install matplotlib seaborn --upgrade --quiet
```

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

**Top 10 most experienced player playing in this years IPL**

```python
final_df.sort_values('Matches', ascending=False).head(10)
```

**Top 10 Players with most Runs scored**

```python
final_df[['Team','Name_Of_Player','Country','Runs_Scored','Style','Role','Innings_Bat']].sort_values('Runs_Scored', ascending=False).head(10)
```

**We will plot a bar chart to compare the runs of top 10 most run scorers**  

```python
top_run_scorer_df=final_df[['Team','Name_Of_Player','Country','Runs_Scored','Style','Role','Innings_Bat']].sort_values('Runs_Scored', ascending=False).head(10)
```

```python

plt.figure(figsize=(10,8))
plt.xticks(rotation=75)
plt.title("Top Run Scorer")
sns.barplot(x='Name_Of_Player', y='Runs_Scored', data=top_run_scorer_df);
```

**Top 10 players having the best Batting averages** 


It will create a wrong assumption if we just sort the batting averages from the original data set since there may be players who has played very few innings and can have a higher averages, so we have chosen a threshold of 20 innings.

```python
new_final_df=final_df[final_df["Innings_Bat"]>20]
```

```python
new_final_df[['Team', 'Name_Of_Player', 'Country', 'Style', 'Role','Bat_Avg', 'Matches',
       'Innings_Bat',  'Runs_Scored', 'Highest_Score',
      ]].sort_values('Bat_Avg', ascending=False).head(10)
```

**Top 10 players having the best Batting Strike Rate** 

For same reason like the Batting averages we have taken players who have played a minimum of 20 innings.

```python
new_final_df[['Team', 'Name_Of_Player', 'Country', 'Style', 'Role', 'Bat_SR','Matches',
       'Innings_Bat',  'Highest_Score',
       'Bat_Avg' ]].sort_values('Bat_SR',ascending=False).head(10)
```

**Top 10 players with most Centuries**

```python
final_df[['Team', 'Name_Of_Player', 'Country', 'Style', 'Role','100s', 'Matches',
       'Innings_Bat', 'Runs_Scored', 'Highest_Score',
       'Bat_Avg', ]].sort_values('100s', ascending=False).head(10)
```

**Top 10 players with most Half-Centuries**

```python
final_df[['Team', 'Name_Of_Player', 'Country', 'Style', 'Role','50s', 'Matches',
       'Innings_Bat', 'Runs_Scored', 'Highest_Score',
       'Bat_Avg', ]].sort_values('50s', ascending=False).head(10)
```

**Top 10 players who have hit the most number of 6s**

```python
final_df[['Team', 'Name_Of_Player', 'Country', 'Style', 'Role','6s', 'Matches',
       'Innings_Bat', 'Runs_Scored', 'Highest_Score',
       'Bat_Avg', ]].sort_values('6s', ascending=False).head(10)
```

**We will see team wise which team has hit the most number of sixes**

We will use the .groupby method to calculate the no of 6s team wise.

```python
final6s_df = final_df.groupby(by=["Team"])[["6s"]].sum()
```

```python
final6s_df.reset_index(inplace=True)
```

```python
final6s_df
```

**We have plotted a bar chart to compare the 6s hit by each team**

```python
sns.barplot(x='Team', y='6s', data=final6s_df);
```
